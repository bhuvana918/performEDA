```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df=df.drop_duplicates()
df.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(df['volatile acidity'],color='green')
```

```python
df.chlorides.value_counts().nlargest(10).plot(kind='bar', figsize=(10,5),color='blue')
plt.xlabel('Chlorides')
plt.ylabel('Wine Quality')
plt.title('Wine Quality on chlorides')
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['citric acid'],color='pink',edgecolor='black')
plt.title('Count of Citric Acid')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['fixed acidity'],y=df['pH'],color='m')
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,annot=True)
```
